```python
xmin, xmax = -np.pi, np.pi
x = np.arange(xmin,xmax,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.title("sss")
plt.hlines(-1,xmin,xmax,linestyle="dashed")

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.subplot(2,1,2)
plt.title("dddd")
plt.plot(x,y_cos)

plt.tight_layout()
```

```python
x = np.arange(-3,3,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig = plt.figure()
print("type(fig): {}".format(type(fig)))

ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
print("type(ax1): {}".format(type(ax1)))

ax1.plot(x,y_sin)
ax2.plot(x,y_cos)

ax1.set_ylim(-1.3,1.3)
ax2.set_ylim(-2,2)

ax1.set_title("sin")
ax2.set_title("cos")

ax1.set_xlabel("x")
ax2.set_xlabel("x")

fig.tight_layout()
```

```python
x = np.random.randn(300)
y1 = np.exp(x) + np.random.randn(300)
y2 = np.exp(x)*3 + np.random.randn(300)

fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122,sharey = ax1)

ax1.plot(x,y1,".")
ax2.plot(x,y2,".")
```
